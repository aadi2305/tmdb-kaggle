```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json 
from sklearn.preprocessing import LabelEncoder
```

```python
df1 = pd.read_csv("tmdb_5000_credits.csv")
df2 = pd.read_csv("tmdb_5000_movies.csv")
```

```python
df2.head()
```

```python
df2 = df2.drop("homepage", axis = 1)
df2 = df2.drop("overview", axis = 1)
df2 = df2.drop("tagline", axis = 1)
```

```python
#adding profit column to df2
profit = []
for i in range(4803):
    profit.append(df2["revenue"][i] -df2["budget"][i])  
df2["Profit"] = profit
```

```python
#for genres
ids1 = []
name1 = []
for i in range(4803):
    ids2 = []
    name2 = []
    contents = json.loads(df2["genres"][i])
    for j in range(len(contents)):
        ids2.append(contents[j]["id"])
        name2.append(contents[j]["name"])
    ids1.append(ids2)
    name1.append(name2)
df2["genre_id"] = ids1
df2["genre_name"] = name1
df2= df2.drop("genres", axis =1)

#for keywords
keyword_id = []
keyword_name = []
for i in range(4803):
    temp_list1 = []
    temp_list2 = []
    contents = json.loads(df2["keywords"][i])
    for j in range(len(contents)):
        temp_list1.append(contents[j]["id"])
        temp_list2.append(contents[j]["name"])
    keyword_id.append(temp_list1)
    keyword_name.append(temp_list2)
df2["keyword_id"] = keyword_id
df2["keyword_name"] = keyword_name
df2= df2.drop("keywords", axis =1)

#for spoken language section
spoken_languages = []
for i in range(4803):
    temp_list = []
    contents = json.loads(df2["spoken_languages"][i])
    for j in range(len(contents)):
        temp_list.append(contents[j]["iso_639_1"])
    spoken_languages.append(temp_list)
df2["spoken_languages"] = spoken_languages

#for produtcion countries section
countries = []
for i in range(4803):
    temp_list = []
    contents = json.loads(df2["production_countries"][i])
    for j in range(len(contents)):
        temp_list.append(contents[j]["name"])
    countries.append(temp_list)
df2["production_countries"] = countries
```

```python
df2 = df2.dropna(axis = 0)
```

```python
date = []
month = []
year = []

for i in range(4800):
    
    if i == 2656 or i ==4140 or i == 4553: #these are the dropped rows which had null values
        year.append(0)
        month.append(0)
        date.append(0)
    else:
        year.append(int(str(df2["release_date"][i]).split("-")[0]))
        month.append(int(str(df2["release_date"][i]).split("-")[1]))
        date.append(int(str(df2["release_date"][i]).split("-")[2]))
df2["date"] = date
df2["month"] = month
df2["year"] = year
df2 = df2.drop("release_date", axis =1)
```

```python
df2 = df2.drop(["title", "original_title"], axis = 1)
```

```python
le = LabelEncoder()
for col in ["original_language", "status"]:
    df2[col] = le.fit_transform(df2[col])
```

```python
#creating dummy varibles for every genre

unique_genre_names = []
for i in range(4803):
    for j in range(len(name1[i])):
        if name1[i][j] in unique_genre_names:
            pass 
        else:
            unique_genre_names.append(name1[i][j])
genre = []
for i in range(4803):
    if i == 2656 or i ==4140 or i == 4553:
        pass
    else:
        k =0
        for genre_name in unique_genre_names:
            genre.append([])
            if genre_name in df2["genre_name"][i]:
                genre[k].append(1)
            else:
                genre[k].append(0)
            k+=1
k = 0
for genre_name in unique_genre_names:
    df2[genre_name] = genre[k]
    k+=1   
    
df2 = df2.drop(["genre_name", "genre_id"], axis =1)
```

```python
#creating dummy varibles for every country

unique_country_names = []
for i in range(4803):
    for j in range(len(countries[i])):
        if countries[i][j] in unique_country_names:
            pass 
        else:
            unique_country_names.append(countries[i][j])
country = []
for i in range(4803):
    if i == 2656 or i ==4140 or i == 4553:
        pass
    else:
        k =0
        for country_name in unique_country_names:
            country.append([])
            if country_name in df2["production_countries"][i]:
                country[k].append(1)
            else:
                country[k].append(0)
            k+=1
k = 0
for country_name in unique_country_names:
    df2[country_name] = country[k]
    k+=1   
    
df2 = df2.drop("production_countries", axis =1)
```
